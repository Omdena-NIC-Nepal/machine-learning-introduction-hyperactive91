# Source: Boston Housing Dataset on UCI Machine Learning Repository

### Features:

- CRIM: per capita crime rate by town.
- ZN: proportion of residential land zoned for lots over 25,000 sq. ft.
- INDUS: proportion of non-retail business acres per town.
- CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise).
- NOX: nitrogen oxides concentration (parts per 10 million).
- RM: average number of rooms per dwelling.
- AGE: proportion of owner-occupied units built prior to 1940.
- DIS: weighted distances to five Boston employment centers.
- RAD: index of accessibility to radial highways.
- TAX: full-value property tax rate per 10000 dollars
- PTRATIO pupil-teacher ratio by town.
- B: (1000(Bk - 0.63)^2) where Bk is the proportion of black residents by town.
- LSTAT: percentage of lower status of the population.
- MEDV: median value of owner-occupied homes in $1000s (Target).

# Visualization between the features and target variable medv

```python
#heat map visualization
fig, ax=plt.subplots(figsize=(10,8))
corr_df= df.corr()
sns.heatmap(data=corr_df,annot=True,cmap="Blues",fmt=".2f") 
print("Correlation between the features and the target variable : \n\n",corr_df['medv'].sort_values(ascending=True))
plt.title("Correlation heatmap")
plt.show()
```

```python
corr_df_target=df.corr()['medv'].sort_values(ascending=False) #sorting the correllation values of features with target variable
plt.figure(figsize=(10,6))
corr_df_target.drop('medv').plot.barh() #dropping the medv(target variable in bar plot)
plt.title("Correlation between features and target variable")
plt.xlabel("Correlation values")
plt.ylabel("Features")
plt.show()
```

# Analysis from the correlation values

### Strongest Positive Correlations with medv:

- rm (0.695360): Average number of rooms per dwelling → More rooms = higher house value.

- zn (0.360445): Proportion of residential land zoned for large lots → More zoning for big homes = higher value.

- b (0.333461): A variable involving proportion of Black residents (the formula is a bit controversial but it's in the dataset).

- dis (0.249929): Distance to employment centers → Surprisingly positive; more distance = higher value, likely due to suburban appeal.

- chas (0.175260): Charles River dummy variable → If the property is near the river, it adds to the value.

### Strongest Negative Correlations with medv:

- lstat (-0.737663): % lower status population → Strongly negative; more low-income residents = lower home value.

- ptratio (-0.507787): Pupil-teacher ratio → Higher ratio = worse schools = lower value.

- indus (-0.483725): Proportion of industrial area → More industry = lower home value.

- tax (-0.468536): Property tax rate → Higher taxes = lower value.

- nox (-0.427321): Air pollution (NOx concentration) → More pollution = lower value.

- crim (-0.388305): Crime rate → More crime = lower value.

- rad (-0.381626): Access to highways → Possibly negative due to noise/traffic.

- age (-0.376955): Age of homes → Older homes = lower value.
